## How did we retrieve the tweets?

We used *TweetedAt* [https://github.com/oduwsdl/tweetedat], a web service and library to extracts date and time from the tweet ID by reverse-engineering Twitter Snowflake.
In particular extracts the timestamps for post-Snowflake IDs directly from it and estimate timestamps for pre-Snowflake IDs.

### What is Snowflake?

In 2010, Twitter migrated its database from MySQL to Cassandra.
<br> Unlike MySQL, Cassandra does not support sequential ID generation technique, so Twitter announced Snowflake, a service to generate unique IDs for all the tweet IDs and other objects within Twitter like lists, users, collections, etc. 
<br>Snowflake generates unsigned-64 bit integers which consist of: 

-   timestamp - 41 bits (millisecond precision w/ a custom epoch gives us 69 years)
-   configured machine ID - 10 bits - gives us up to 1024 machines
-   sequence number - 12 bits - rolls over every 4096 per machine (with protection to avoid rollover in the same ms)

#### Why not check on Twitter directly?

-   Also Twitter developer API has access rate limits. It acts as a bottleneck in finding timestamps over a data set of tweet IDs. This bottleneck is not present in TweetedAt because we do not interact with Twitter's developer API for finding them.
-   Deleted, suspended, and protected tweets do not have their metadata accessible from Twitter's developer API. TweetedAt is the solution for finding the timestamps of any of these inaccessible tweets.


### How is the timestampo obtained?

**Post SnowFlake ID**

According to the official documentation about snowflake the tweet IDs are k-sorted within a second bound but the millisecond bound cannot be guaranteed.
We can extract the timestamp for a tweet ID by right shifting the tweet ID by 22 bits and adding the Twitter epoch time of 1288834974657.  

    def get_tweet_timestamp(tid):

Pre-Snowflake IDs do not have their timestamps encoded in the IDs, but we can uncover the value from the 2362 tweet IDs with known timestamps. 

For TweetedAt implementation, 2362 tweet IDs and their timestamps were collected at a daily interval between March 2006 and November 2010 to create a ground truth data set. The ground truth data set is used for estimating the timestamp given the TweetID. Using the weekly interval as ground truth data set resulted in an approximate error of 1 hour.

```python
import os
import pandas as pd

def removeUserMention(string):
    if string.startswith('@'):
        return "__user_mention__"
    return string

def cleanMentions(tweet):
    try:
        tokens = tweet.split()
        tokens = [removeUserMention(elem) for elem in tokens]  
        return ' '.join([str(elem) for elem in tokens])
    except:
        return tweet

dir_list = os.listdir("dataset/added")

def removeUrl(string):
    if string.startswith('http'):
        return "__url__"
    return string

def cleanUrls(tweet):
    try:
        tokens = tweet.split()
        tokens = [removeUrl(elem) for elem in tokens]  
        return ' '.join([str(elem) for elem in tokens])
    except:
        return tweet

start = pd.read_csv("dataset/fixed.csv")
users = start[['screen_name_anonymized', 'screen_name', 'account.type', 'user_id']]
first = True
monthly_df = []

for i in range(0, len(dir_list)):
    temp = pd.read_csv("dataset/added/"+dir_list[i])
    temp.rename(columns = {'tweet':'Tweet','id':'status_id', 'username':'screen_name'}, inplace = True)
    temp = temp[["status_id", "date", "time", "timezone", "user_id", "screen_name", "Tweet"]]
    pd.options.mode.chained_assignment = None  

    # tag removal
    temp['Tweet'] = temp['Tweet'].apply(cleanMentions)
    

    # url removal
    temp['Tweet'] = temp['Tweet'].apply(cleanUrls)

    #metto la classe
    user_map = {}
    for x in users['screen_name'].unique():
        user_map[x] = users[users['screen_name'] == x].iloc[0]['account.type']
    temp['Account_type'] = None
    temp['Account_type'] = temp['screen_name'].apply(lambda x : user_map[x] if x in users['screen_name'].unique() else None)
    if first:
        tweet = pd.DataFrame(columns=temp.columns)
        first = False
    tweet = pd.concat([tweet,temp], ignore_index = True)
    monthly_df.append(temp)
```

**Incremental**: the classification model is updated based on the new labeled chunk of tweets with a partial fitting. 

Only the classifier is updated, whereasthe attribute space, namely the vocabulary generated based on the initialTF-IDF vectorization, is left unchanged throughout the online monitoring.

We are forced to use Native Bayesian classifier

```python
from sklearn.naive_bayes import MultinomialNB
pipeline_incremental = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words=stopwords, max_df=0.22, min_df=0.0012)),
    ('clf', MultinomialNB())
])   
pipeline_incremental.fit(dataset.Tweet,dataset.Account_type)
```

```python
def partial_pipe_fit(pipeline_obj, X,y):
    X_transformed = pipeline_obj.named_steps['tfidf'].transform(X)
    sm = SMOTE(random_state=123, k_neighbors = 5)
    X_res, y_res = sm.fit_resample(X_transformed, y)
    pipeline_obj.named_steps['clf'].partial_fit(X_res,y_res)
```

```python
results_incremental = []
f1_incremental = []
results_incremental.append(pipeline_incremental.score(monthly_df_cleaned[0].Tweet,monthly_df_cleaned[0].Account_type))
results_incremental.append(pipeline_incremental.score(monthly_df_cleaned[1].Tweet,monthly_df_cleaned[1].Account_type))
y_pred = pipeline_incremental.predict(monthly_df_cleaned[0].Tweet)
f1_incremental.append(f1_score(monthly_df_cleaned[0].Account_type, y_pred, average='macro'))
y_pred = pipeline_incremental.predict(monthly_df_cleaned[1].Tweet)
f1_incremental.append(f1_score(monthly_df_cleaned[1].Account_type, y_pred, average='macro'))

for i in range(2,len(monthly_df_cleaned)):
    partial_training_set = pd.concat([monthly_df_cleaned[i-1],monthly_df_cleaned[i-2]])
    partial_pipe_fit(pipeline_incremental,partial_training_set.Tweet,partial_training_set.Account_type)
    results_incremental.append(pipeline_incremental.score(monthly_df_cleaned[i].Tweet,monthly_df_cleaned[i].Account_type))
    y_pred = pipeline_incremental.predict(monthly_df_cleaned[i].Tweet)
    f1_incremental.append(f1_score(monthly_df_cleaned[i].Account_type, y_pred, average='macro'))
```
